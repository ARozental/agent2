```python
def is_empty(a):
    if a==[0]:
        return True
    if type(a)!=list:
        return False
    elif len(a)>1:
        return False
    elif len(a)==0:
        return True
    elif type(a[0])!=list:
        return False
    else:
        return is_empty(a[0])
    
def remove_empty(a):
    if type(a)!=list:
        return a  
    return [remove_empty(x) for x in a if is_empty(x)==False]
    
def get_depth(nested,res=-1):
    if type(nested)==int:
        return res
    else:
        return get_depth(nested[0],res+1)
    
def split_text(text,tokenize_fn, current_level):
    delimiters = ['', ' ', '  ', '\n','Chapter ']  # ,words,sentences,paragraphs
    if current_level == 0:
        return tokenize_fn(text)
    f = lambda text: text.split(delimiters[current_level])
    return [split_text(t,tokenize_fn, current_level - 1) for t in f(text)]
```

```python
def get_depth(nested,res=-1):
    try:
        if type(nested)==int:
            return res
        return get_depth(nested[0],res+1)
    except:
        print(nested)
        return 0

def join_text(nested,reverse_tokenizer_fn):
    current_level = get_depth(nested)
    delimiters = ['', ' ', '  ', '\n','\n\n']  # letters,words,sentences,paragraphs
    f = lambda text: delimiters[current_level].join(text)
    if current_level == 0:
        res = reverse_tokenizer_fn(nested)
        return f(res)
    return f([join_text(arr,reverse_tokenizer_fn) for arr in nested])
    
join_text([[[[45, 58, 48],[45, 58, 48, 0]],[[45, 58, 48, 0],[45, 58, 48, 0]]],[[[45, 58, 48, 0],[45, 58, 48, 0]],[[45, 58, 48, 0],[45, 58, 48, 0]]]],create_reverse_tokenizer("../chars.txt"))
```

```python
from collections import defaultdict
import glob

def create_tokenizer(char_file):  # "../chars.txt"
  x = open(char_file, "r").readlines()
  tokenizer = defaultdict(int, dict(zip([l.strip() for l in x], range(1, 7777))))
  def tokenize(word):
    return [tokenizer[l] for l in word]+[tokenizer['end-of-word-token']]
  return tokenize

def create_reverse_tokenizer(char_file):  # "../chars.txt"
  x = open(char_file, "r").readlines()
  reverse_tokenizer = dict(zip(range(1, 500), [l.strip() for l in x]))
  reverse_tokenizer[0] = ''

  def tokenize(token_list):
    return [reverse_tokenizer[t] for t in token_list]

  return tokenize


delimiters = ['', ' ', '  ', '\n','\n\n']  # letters,words,sentences,paragraphs,chapters (though it is "Chapter " really)
def split_text(text,tokenize_fn, current_level):
  if current_level == 0:
    return tokenize_fn(text)
  f = lambda text: text.split(delimiters[current_level])
  return [split_text(t,tokenize_fn, current_level - 1) for t in f(text)]


def is_empty(a):
  if a == [0]:
    return True
  if type(a) != list:
    return False
  elif len(a) > 1:
    return False
  elif len(a) == 0:
    return True
  elif type(a[0]) != list:
    return False
  else:
    return is_empty(a[0])

def remove_empty(a):
  if type(a) != list:
    return a
  return [remove_empty(x) for x in a if is_empty(x) == False]

def do_book(location):
  tokenize_fn = create_tokenizer("../chars.txt")
  lines = open(location, "r").readlines()
  book_title = lines[10].strip()
  text = "".join(lines)
  chapters = text.split("Chapter ")
  chapters = [split_text("Chapter " + x,tokenize_fn, 3) for x in chapters]
  chapters[0] = split_text(book_title,tokenize_fn, 3)
  return remove_empty(chapters)

def get_depth(nested, res=-1):
  if type(nested) == int:
    return res
  return get_depth(nested[0], res + 1)


def join_text(nested, reverse_tokenizer_fn):
  current_level = get_depth(nested)
  f = lambda text: delimiters[current_level].join(text)
  if current_level == 0:
    res = reverse_tokenizer_fn(nested)
    return f(res)
  return f([join_text(arr, reverse_tokenizer_fn) for arr in nested])
print(do_book("../datasets/Scholar_of_Magic_2_trial.txt"))
```

```python
import tensorflow as tf
import glob


class Dataset(object):
    def __init__(self):
        self.next_element = self.build_iterator()
    def get_next_book(self):
        all_books = glob.glob("../datasets/*.txt")
        l = len(all_books)
        num = 0
        while True:
            out = do_book(all_books[num%l])
            ragged_out =tf.ragged.constant(out)
            yield ({
              "flat_values":ragged_out.flat_values,
              "nested_row_splits": ragged_out.nested_row_splits
              })
            num += 1
            
    def build_iterator(self):

        batch_size = 1
        prefetch_batch_buffer = 0
        dataset = tf.data.Dataset.from_generator(self.get_next_book,output_types = {
                                                                               "flat_values":tf.int64, "nested_row_splits":(tf.int64,tf.int64,tf.int64,tf.int64)
                                                                               })

        dataset = dataset.batch(batch_size)
        dataset = dataset.prefetch(prefetch_batch_buffer)
        it = tf.compat.v1.data.make_one_shot_iterator(dataset)

        return it
ds = Dataset()
ds.next_element.get_next()
#print(ds.next_element.get_next())
```

```python
def max_lengths(book):
    chapters = book
    l1 = len(chapters)
    l2 = max([len(x) for x in chapters])
    paragraphs = [item for sublist in chapters for item in sublist]
    l3 = max([len(x) for x in paragraphs])
    sentences =[item for sublist in paragraphs for item in sublist]
    l4 = max([len(x) for x in sentences])
    words =[item for sublist in sentences for item in sublist]
    l5 = max([len(x) for x in words])

    return [l5,l4,l3,l2,l1,1]

max_lengths(do_book("../datasets/Scholar_of_Magic_2_trial.txt")),max_lengths(do_book("../datasets/Scholar_of_Magic_3_trial.txt"))
```

```python
import numpy as np
def create_segment_b_mask(sequence_length,dtype):
  #p=np.round(np.random.uniform())*np.random.uniform()
  p=tf.round(tf.random.uniform([],dtype=dtype))*tf.random.uniform([],dtype=dtype)
  if p==0:
    return tf.expand_dims(np.zeros(shape=[sequence_length],dtype=dtype),axis=-1)
  r = np.floor(np.random.uniform(size=[sequence_length])/(1.0-p))
  #r = tf.floor(tf.random.uniform([sequence_length],dtype=dtype)/(1.0-p))

  segment_b = tf.cast(np.divide(r,r,out=r, where=r>0),dtype=dtype)
  return  tf.expand_dims(segment_b * np.round(np.random.uniform(size=[sequence_length])),axis=-1)
```
